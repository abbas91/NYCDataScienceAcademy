#### (1) Use three variables 'Height', 'Length' and 'Rings' from abalone dataset to make a scatter matrix like this:
<img src=http://slides.nycdatascience.com/BOOTCAMP3/content/week6/pic/scatter_matrix.png>

```python
import pandas as pd
abalone = pd.read_csv('./abalone.data', header = 0,
                      names = ['Sex', 'Length', 'Diam', 'Height', 'Whole', 'Shucked', 'Viscera', 'Shell', 'Rings'])
```

```python
abalone.head()
```

```python
Height = abalone.Height
Length = abalone.Length
Rings = abalone.Rings
```

- Figure size: width 12; height 12.

- Transparency: 0.5

- Pay attention to the labels on the axes.

```python
### write your code here
%matplotlib inline
import matplotlib.pyplot as plt
n = 3
data = [Height, Length, Rings]
labels = ['Height', 'Length', 'Rings']
plt.rcParams['figure.figsize'] = 12, 12
for i in range(3):
    for j in range(3):
        # position
        plt.subplot2grid((3, 3), (i, j))
        # hist & scatter
        if i == j:
            plt.hist(data[i], alpha = 0.5)
        else:
            plt.scatter(data[i], data[j], alpha = 0.5)
        # xlabel & ylabel
        if j == 0:
            plt.ylabel(labels[i])
        if i == 2:
            plt.xlabel(labels[j])
```

```python
mtcars = pd.read_csv("./mtcars.csv")
mtcars.head()
```

```python
### write your codes here
g =sns.FacetGrid(mtcars, hue='gear', size=8)
g.map(sns.kdeplot, 'mpg', shade=True)
g.add_legend()
plt.title('Distribution of Gas Milage', fontsize=20)
plt.xlabel( "Miles Per Gallon(mpg)", fontsize=15)
plt.ylabel( "density", fontsize=15)
```
