### Problem 1

Use three variables 'Height', 'Length' and 'Rings' from abalone dataset to make a scatter matrix like this:
<img src=http://slides.nycdatascience.com/BOOTCAMP3/content/week6/pic/scatter_matrix.png>

```python
import pandas as pd
abalone = pd.read_csv('./abalone.data', header = 0,
                      names = ['Sex', 'Length', 'Diam', 'Height', 'Whole', 'Shucked', 'Viscera', 'Shell', 'Rings'])
```

```python
abalone.head()
```

```python
Height = abalone.Height
Length = abalone.Length
Rings = abalone.Rings
```

- Figure size: width 12; height 12.

- Transparency: 0.5

- Pay attention to the labels on the axes.

```python
### write your code here
%pylab inline
import matplotlib.pyplot as plt
%matplotlib inline

rcParams['figure.figsize'] = 12, 12
plt.style.use('seaborn-ticks')

alpha_hist = 0.5
color = 'mediumblue'

plt.figure(1)

plt.subplot(331)
plt.hist(Height, alpha = alpha_hist)
plt.ylabel('Height')

plt.subplot(332)
plt.scatter(Length, Height)

plt.subplot(333)
plt.scatter(Rings, Height)

plt.subplot(334)
plt.scatter(Height, Length)
plt.ylabel('Length')

plt.subplot(335)
plt.hist(Length, alpha = alpha_hist)

plt.subplot(336)
plt.scatter(Rings, Length)

plt.subplot(337)
plt.scatter(Height,Rings)
plt.ylabel('Rings')
plt.xlabel('Height')

plt.subplot(338)
plt.scatter(Length,Rings)
plt.xlabel('Length')

plt.subplot(339)
plt.hist(Rings,  alpha=alpha_hist)
plt.xlabel('Rings')

plt.show()
```
